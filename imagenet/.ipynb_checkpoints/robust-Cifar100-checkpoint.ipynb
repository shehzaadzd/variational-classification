```python
import numpy as np
from evaluation import Evaluator
from scipy.special import softmax
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
corruptions = ["brightness","defocus_blur", "fog","gaussian_noise","impulse_noise","motion_blur","shot_noise","zoom_blur","contrast","elastic_transform","glass_blur","jpeg_compression","pixelate","snow"]

labels = np.load("classes.npy")
evaluator = Evaluator()
```

```python
accuracy_CE = np.zeros([5, len(corruptions)])
ECEs_CE =np.zeros([5, len(corruptions)])

accuracy_CE_t = np.zeros([5, len(corruptions)])
ECEs_CE_t =np.zeros([5, len(corruptions)])

accuracy_AVB = np.zeros([5, len(corruptions)])
ECEs_AVB = np.zeros([5, len(corruptions)])
```

```python
for j, corruption in enumerate(["brightness","defocus_blur", "fog","gaussian_noise","impulse_noise","motion_blur","shot_noise","zoom_blur","contrast","elastic_transform","glass_blur","jpeg_compression","pixelate","snow"]):
    for i, lev in enumerate(["1", "2", "3", "4", "5"]):
        file_name = "outputs/output_tiny_AVB_{}_{}.npy".format(corruption,lev) 
        op = np.load(file_name)
        probs = softmax(op, axis=-1)
        preds = np.argmax(probs, axis=-1)
        confs = np.max(probs, axis=-1)
        accs = preds == labels
        ECE, MCE, bin_accuracy, bin_confidence, _ = evaluator.create_fine_bins(accs, confs)
        accuracy_CE[i][j] = np.mean(accs)
        ECEs_CE[i][j] = ECE
        
        
        
```

```python
for j, corruption in enumerate(["brightness","defocus_blur", "fog","gaussian_noise","impulse_noise","motion_blur","shot_noise","zoom_blur","contrast","elastic_transform","glass_blur","jpeg_compression","pixelate","snow"]):
    for i, lev in enumerate(["1", "2", "3", "4", "5"]):
        file_name = "outputs/output_tiny_AVB_{}_{}.npy".format(corruption,lev) 
        op = np.load(file_name)
        probs = softmax(op/1.12, axis=-1)
        preds = np.argmax(probs, axis=-1)
        confs = np.max(probs, axis=-1)
        accs = preds == labels
        ECE, MCE, bin_accuracy, bin_confidence, _ = evaluator.create_fine_bins(accs, confs)
#         accuracy_CE[i][j] = np.mean(accs)
        ECEs_CE_t[i][j] = ECE
        
```

```python
for j, corruption in enumerate(["brightness","defocus_blur", "fog","gaussian_noise","impulse_noise","motion_blur","shot_noise","zoom_blur","contrast","elastic_transform","glass_blur","jpeg_compression","pixelate","snow"]):
    for i, lev in enumerate(["1", "2", "3", "4", "5"]):
        file_name = "outputs/output_tiny_VC_{}_{}.npy".format(corruption,lev) 
        op = np.load(file_name)
        probs = softmax(op, axis=-1)
        preds = np.argmax(probs, axis=-1)
        confs = np.max(probs, axis=-1)
        accs = preds == labels
        ECE, MCE, bin_accuracy, bin_confidence, _ = evaluator.create_fine_bins(accs, confs)
        accuracy_AVB[i][j] = np.mean(accs)
        ECEs_AVB[i][j] = ECE
        
```

```python
a = np.concatenate([ECEs_CE[:,:,np.newaxis], ECEs_CE_t[:,:,np.newaxis], ECEs_AVB[:,:,np.newaxis]], axis=-1 )
m,n,r = a.shape
out_arr = np.column_stack((np.repeat(np.arange(m),n),a.reshape(m*n,-1)))
new_out_arr= np.stack([np.concatenate([[0]*70, [1]*70, [2]*70]), np.concatenate([out_arr[:, 0],]*3, axis=0), np.concatenate([out_arr[:, 1], out_arr[:, 2], out_arr[:, 3]])])
out_df = pd.DataFrame(new_out_arr.T)
out_df.loc[out_df[0] == 2, 0] = "VC"
out_df.loc[out_df[0] == 1, 0] = "CE temp"
out_df.loc[out_df[0] == 0, 0] = "CE"
out_df.loc[out_df[1]>= 0.0, 1] = out_df[1] + 1
out_df = out_df.rename(columns={0: "Model", 1: "Shift Intensity", 2:"ECE"})

sns.set_theme()
# sns.violinplot(accuracy_CE.T )
sns_plot = sns.boxplot(data=out_df, x="Shift Intensity", y="ECE", hue="Model", showfliers = False)
sns_plot.figure.savefig("cifar_100_c_ece.pdf")
```
